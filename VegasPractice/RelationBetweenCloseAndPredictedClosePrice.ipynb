```scala
import $ivy.`org.vegas-viz::vegas:0.3.11`

import $ivy.`org.vegas-viz::vegas-spark:0.3.11`

import $ivy.`org.apache.spark::spark-sql:2.4.0`
import org.apache.spark.SparkContext._

import org.apache.spark.rdd._


import org.apache.log4j._

import org.apache.spark.sql._
import vegas._

import vegas.render.WindowRenderer._

import vegas.sparkExt._
```

```scala
val sparkSession = SparkSession.builder.appName("Stock Prediction Analysis").master("local[*]").getOrCreate()
Logger.getLogger("org").setLevel(Level.ERROR)
```

```scala
val inputDataFrame = sparkSession.read.option("inferSchema",true).option("header",true).csv("./SavedOutput/*.csv")
```

```scala
inputDataFrame.show(5)
```

```scala
Vegas("Change of Predicted Close Price with respect to Time")
    .withDataFrame(inputDataFrame)
    .mark(Line)
    .encodeX("Date",Ordinal,axis = Axis(grid = true))
    .encodeY("Predicted Close Price",Quant,scale = Scale(domainValues = List(1600, 1650.0)))
    .show
    
```

```scala
Vegas("Relation Between Predicted Close Price And Close")
    .withDataFrame(inputDataFrame)
    .mark(Line)
    .encodeX("Predicted Close Price",Ordinal,axis = Axis(grid = true))
    .encodeY("Close",scale = Scale(domainValues = List(1600, 1650.0)))
    .show
    
```

```scala
Vegas.layered("Relation Between Predicted Close Price And Close")
    .withDataFrame(inputDataFrame)
    .withLayers(
        Layer()
        .mark(Line)
        .encodeX("Date",Ordinal,axis = Axis(grid = true))
        .encodeY("Close",Quant,scale = Scale(domainValues = List(1630, 1650.0)),axis = Axis(title = "Close Price"))
        .configMark(color = "Green"),
        Layer()
        .mark(Line)
        .encodeX("Date",Ordinal,axis = Axis(grid = true))
        .encodeY("Predicted Close Price",Quant,scale = Scale(domainValues = List(1630, 1650.0)))
        .configMark(color="red")
        )
    .show
    
```
